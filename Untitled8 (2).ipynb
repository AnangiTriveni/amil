```python
Observations from correlations plots and coeffcients

# Between x and y all the variables are showing moderate to high correlation strengths,highest being between HP and MPG
# There this dataset qualifies for building linear regreesion model to predict MPG
# Among x columns(x1,x2,x3 and x4),some very high correlation strengths SP vs HP,VOL vs WT
```

```python
model1 = smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
```

```python
model1.summary()
```

```python
Observations from model summary 

# The R-squared and adjusted R-squared values are good and about 75% of variability in Y is explanied by X columns
# The probability value with respect to F-statistic is close to zero,indicating that all or someod X columns are significant 
# The p-values for VOL and WT are higher than 5% indicating issue among themseleves which need to explored
```

```python
df1 = pd.DataFrame()
df1["actual_y1"] = cars["MPG"]
df1.head()
```

```python
pred_y1 = model1.predict(cars.iloc[:,0:4])
df1["pred_y1"] = pred_y1
df1.head()
```
